# **Identifying the imbalance data**

```python
not_remake_films = films[films['Whether Remake'] == 'No'].shape[0]
remake_films = films[films['Whether Remake'] == 'Yes'].shape[0]
print("Ratio is :- ", not_remake_films, ":" ,remake_films)
px.bar(films['Whether Remake'].value_counts(), films['Whether Remake'].value_counts(), labels = {'x' : 'Number of values',
                                                                                                 'index' : 'Whether Remake'}, 
       title = 'Data of remake is imbalance')
```

```python
not_franchise_films = films[films['Whether Franchise'] == 'No'].shape[0]
franchise_films = films[films['Whether Franchise'] == 'Yes'].shape[0]
print("Ratio is :- ", not_franchise_films, ":" ,franchise_films)
px.bar(films['Whether Franchise'].value_counts(), films['Whether Franchise'].value_counts(), labels = {'x' : 'Number of values',
                                                                                                 'index' : 'Whether Franchise'}, 
       title = 'Data of Franchise is imbalance')
```

```python
not_new_actor = films[films['New Actor'] == 'No'].shape[0]
new_actor = films[films['New Actor'] == 'Yes'].shape[0]
print("Ratio is :- ", not_new_actor, ":" ,new_actor)
px.bar(films['New Actor'].value_counts(), films['New Actor'].value_counts(), labels = {'x' : 'Number of values',
                                                                                                 'index' : 'New Actor'}, 
       title = 'Data of New Actor is no so imbalance')
```

```python
not_new_director = films[films['New Director'] == 'No'].shape[0]
new_director = films[films['New Director'] == 'Yes'].shape[0]
print("Ratio is :- ", not_new_director, ":" ,new_director)
px.bar(films['New Director'].value_counts(), films['New Director'].value_counts(), labels = {'x' : 'Number of values',
                                                                                                 'index' : 'New Director'}, 
       title = 'Data of New Director is balance')
```

```python
not_new_music_director = films[films['New Music Director'] == 'No'].shape[0]
new_music_director = films[films['New Music Director'] == 'Yes'].shape[0]
print("Ratio is :- ", not_new_music_director, ":" ,new_music_director)
px.bar(films['New Music Director'].value_counts(), films['New Music Director'].value_counts(), labels = {'x' : 'Number of values',
                                                                                                 'index' : 'New Music Director'}, 
       title = 'Data of New Music Director is not so imbalance')
```

```python
not_holiday_release = films[films['Release Period'] == 'Normal'].shape[0]
holiday_release = films[films['Release Period'] == 'Holiday'].shape[0]
print("Ratio is :- ", not_holiday_release, ":" ,holiday_release)
px.bar(films['Release Period'].value_counts(), films['Release Period'].value_counts(), labels = {'x' : 'Number of values',
                                                                                                 'index' : 'Release Period'}, 
       title = 'Data of Release Period is not so imbalance')
```
